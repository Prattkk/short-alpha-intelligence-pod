```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
from scipy import stats

plt.style.use('seaborn-v0_8-whitegrid')
pd.set_option('display.max_columns', None)
```

```python
# Load sample data
df = pd.read_csv('../data/sample_short_interest.csv')
df['Business Date'] = pd.to_datetime(df['Business Date'])
print(f'Rows: {len(df)} | Tickers: {df["Ticker"].nunique()}')
df.head()
```

```python
# Summary statistics by ticker
df.groupby('Ticker')[['Squeeze Score', 'Crowded Score', 'ShortInterestPct']].describe().round(2)
```

```python
# Squeeze Score distribution by ticker
fig, axes = plt.subplots(1, 5, figsize=(18, 4), sharey=True)
for i, ticker in enumerate(df['Ticker'].unique()):
    sub = df[df['Ticker'] == ticker]
    axes[i].bar(sub['Business Date'], sub['Squeeze Score'], color='steelblue', alpha=0.8)
    axes[i].set_title(ticker, fontweight='bold')
    axes[i].tick_params(axis='x', rotation=45)
axes[0].set_ylabel('Squeeze Score')
plt.suptitle('Squeeze Score by Ticker at Peak Windows', fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Short Interest % vs Squeeze Score scatter
fig = px.scatter(
    df, x='ShortInterestPct', y='Squeeze Score',
    color='Ticker', size='Crowded Score',
    title='Short Interest % vs Squeeze Score',
    labels={'ShortInterestPct': 'Short Interest (% Float)', 'Squeeze Score': 'Squeeze Score'},
)
fig.show()
```

## Key Observations

- **AFRM** peaks at 97.5 Squeeze Score — the highest in the dataset
- **SQ** has consistently high Crowded Scores (~47-50) suggesting chronic crowding
- **TSLA** shows the widest date spread, with squeeze events across both Jan and Nov 2021
- Short Interest % alone doesn't predict squeeze intensity — SQ has the highest SI% but not the highest Squeeze Score
